```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import os
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller, kpss
from dateutil.parser import parse
```

```python

# Import Data
df = pd.read_csv('../data/variables/data-colombia-month-fill.zip', parse_dates=['fecha'], index_col='fecha')
df.head()
```

```python
def decomposition(varname):
    print(varname)
    
    # Additive Decomposition
    result_add = seasonal_decompose(df[varname], model='additive', extrapolate_trend='freq')

    # Plot
    plt.figure(facecolor='w') 
    plt.rcParams.update({'figure.figsize': (10,10)})
    result_add.plot().suptitle('', fontsize=22)
    #plt.show()
    p = 'images/decomposition/'+varname+'.png'
    plt.savefig(p)


    #test for stationarity

    # ADF Test: null hypothesis is the time series possesses a unit root and is non-stationary
    result = adfuller(df[varname], autolag='AIC')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    for key, value in result[4].items():
        print('Critial Values:')
        print(f'   {key}, {value}')

    # KPSS Test: The KPSS test is used to test for trend stationarity. The null hypothesis and the P-Value interpretation is the opposite of ADH test. 
    result = kpss(df[varname], regression='c')
    print('\nKPSS Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    for key, value in result[3].items():
        print('Critial Values:')
        print(f'   {key}, {value}')
```

```python
df.columns
```

```python
for varname in df.columns[2:7]:
    decomposition(varname = varname)
```

## Analysis per basin


```python

# Import Data
df_all = pd.read_csv('../data/variables/data-cuenca-month-fill.zip', parse_dates=['fecha'], index_col='fecha')
df_all.head()
```

```python
df_all.columns
```

```python
def decomposition_basin(df, varname, basin):
    print(varname)
    
    # Additive Decomposition
    result_add = seasonal_decompose(df[df.cuenca==basin][varname], model='additive', extrapolate_trend='freq')

    # Plot
    plt.figure(facecolor='w') 
    plt.rcParams.update({'figure.figsize': (10,10)})
    result_add.plot().suptitle('', fontsize=22)
    #plt.show()
    
    p = 'images/decomposition-basin/'+str(basin)+'/'+varname+'.png'
    plt.savefig(p)


    #test for stationarity

    # ADF Test: null hypothesis is the time series possesses a unit root and is non-stationary
    result = adfuller(df[df.cuenca==basin][varname], autolag='AIC')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    for key, value in result[4].items():
        print('Critial Values:')
        print(f'   {key}, {value}')

    # KPSS Test: The KPSS test is used to test for trend stationarity. The null hypothesis and the P-Value interpretation is the opposite of ADH test. 
    result = kpss(df[df.cuenca==basin][varname], regression='c')
    print('\nKPSS Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    for key, value in result[3].items():
        print('Critial Values:')
        print(f'   {key}, {value}')
```

```python
basin=27
os.mkdir('images/decomposition-basin/'+str(basin))
for varname in df_all.columns[1:6]:
    decomposition_basin(df=df_all, varname = varname, basin = basin)
```
