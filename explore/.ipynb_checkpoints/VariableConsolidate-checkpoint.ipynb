## Libraries

```python
import pandas as pd

pd.options.mode.chained_assignment = None  # default='warn'
```

------

## Data sources 

```python
temperature_mean = pd.read_csv('../data/matrix/cuenca_T_MEDIA_CON_M.zip')
```

```python
rainfall_total = pd.read_csv('../data/matrix/cuencaCANT_PTPM_CON_M.zip')
```

```python
forest_loss_cover = pd.read_csv('../data/matrix/FCL316Basins_month.zip')
```

```python
flow_mean = pd.read_csv('../data/matrix/cuenca_Q_MEDIA_M.zip')
```

------

## Data sources basic descriptions, transformatios and new variables

### Temperature mean

```python
print(temperature_mean.shape)
print(temperature_mean.dtypes)
temperature_mean.head(10)
```

```python
temp_df = temperature_mean[['cuenca', 'codigo', 'Fecha', 'TMEDIA_CO']]
temp_df.columns = ['mc', 'station_id', 'date', 'v_temperature_mean']

temp_df['year'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[0]) , axis = 1) 
temp_df['month'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[1]) , axis = 1) 

#temp_df.drop(columns = ['date'], axis = 1, inplace = True)

temp_df = temp_df[['date', 'year', 'month', 'mc', 'station_id', 'v_temperature_mean']]

temperature_mean = temp_df.copy()

print(temperature_mean.shape)
print(temperature_mean.dtypes)
temperature_mean.head(10)
```

### Rainfall total

```python
print(rainfall_total.shape)
print(rainfall_total.dtypes)
rainfall_total.head(10)
```

```python
temp_df = rainfall_total[['CUENCA', 'Fecha', 'PTPM_CON_WA']]
temp_df.columns = ['mc', 'date', 'v_rainfall_total']

temp_df['year'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[0]) , axis = 1) 
temp_df['month'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[1]) , axis = 1) 

#temp_df.drop(columns = ['date'], axis = 1, inplace = True)

temp_df['station_id'] = None

temp_df = temp_df[['date', 'year', 'month', 'mc', 'station_id', 'v_rainfall_total']]

rainfall_total = temp_df.copy()

print(rainfall_total.shape)
print(rainfall_total.dtypes)
rainfall_total.head(10)
```

### Forest Loss Cover

```python
print(forest_loss_cover.shape)
print(forest_loss_cover.dtypes)
forest_loss_cover.head()
```

```python
temp_df = forest_loss_cover[['year', 'month', 'mc', 'loss_cover']]
temp_df.columns = ['year', 'month', 'mc', 'v_loss_cover']

temp_df['station_id'] = None
temp_df['date'] = temp_df['year'].astype(str).str.pad(width = 2, side = 'left', 
                                                      fillchar = '0') + \
                  '-' + temp_df['month'].astype(str).str.pad(width = 2, side = 'left', 
                                                             fillchar = '0')

temp_df = temp_df[['date', 'year', 'month', 'mc', 'station_id', 'v_loss_cover']]

temp_df.columns = ['date', 'year', 'month', 'mc', 'station_id', 'v_loss_cover']

forest_loss_cover = temp_df.copy().reset_index()

print(forest_loss_cover.shape)
print(forest_loss_cover.dtypes)
forest_loss_cover.tail(10)
```

### Flow mean

```python
print(flow_mean.shape)
print(flow_mean.dtypes)
flow_mean.head(10)
```

```python
temp_df = flow_mean[['cuenca', 'codigo', 'Fecha', 'Q_MEDIA_D']]
temp_df.columns = ['mc', 'station_id', 'date', 'v_flow_mean']

temp_df['year'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[0]) , axis = 1) 
temp_df['month'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[1]) , axis = 1) 

#temp_df.drop(columns = ['date'], axis = 1, inplace = True)

temp_df = temp_df[['date', 'year', 'month', 'mc', 'station_id', 'v_flow_mean']]

flow_mean = temp_df.copy()

print(flow_mean.shape)
print(flow_mean.dtypes)
flow_mean.head(10)
```

------

## Consolidate

```python
matrix_consol = forest_loss_cover.copy()

matrix_consol = matrix_consol.merge(rainfall_total, 
                                    left_on = ['year', 'month', 'mc'], 
                                    right_on = ['year', 'month', 'mc'], 
                                    how = 'left' )

matrix_consol.drop(columns = ['date_y'], axis = 1, inplace = True)

matrix_consol = matrix_consol.merge(temperature_mean, 
                                    left_on = ['year', 'month', 'mc'], 
                                    right_on = ['year', 'month', 'mc'], 
                                    how = 'left' )

matrix_consol.drop(columns = ['date'], axis = 1, inplace = True)

matrix_consol = matrix_consol.merge(flow_mean, 
                                    left_on = ['year', 'month', 'mc'], 
                                    right_on = ['year', 'month', 'mc'], 
                                    how = 'left' )

matrix_consol.drop(columns = ['station_id_y', 'station_id_x', 'date'], axis = 1, 
                   inplace = True)

matrix_consol = matrix_consol[['date_x', 'year', 'month', 'mc', 'v_flow_mean', 
                               'v_loss_cover', 'v_rainfall_total', 
                               'v_temperature_mean']].reset_index(drop = True)

matrix_consol.columns = ['date', 'year', 'month', 'mc', 'v_flow_mean', 'v_loss_cover', 
                         'v_rainfall_total', 'v_temperature_mean']

print(matrix_consol.dtypes)
print(matrix_consol.shape)
matrix_consol.tail()
```

```python
matrix_consol.to_csv('../temp/matrix_consol.csv', index = False)
```

```python
matrix_consol.to_csv('../data/matrix/matrix_consol.zip', index = False, 
                     compression = 'zip')
```
