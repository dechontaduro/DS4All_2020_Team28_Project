## Libraries

```python
import pandas as pd

pd.options.mode.chained_assignment = None  # default='warn'
```

------

## Data sources 

```python
temperature_mean = pd.read_csv('../data/matrix/cuenca_T_MEDIA_CON_M.zip')
```

```python
rainfall_total = pd.read_csv('../data/matrix/cuenca_PTPM_CON_M.zip')
```

```python
forest_loss_cover = pd.read_csv('../data/matrix/cuenca_FORESTCOVERLOSS_M.zip')
```

```python
flow_mean = pd.read_csv('../data/matrix/cuenca_Q_MEDIA_M.zip')
```

------

## Data sources basic descriptions, transformatios and new variables

### Temperature mean

```python
print(temperature_mean.dtypes)
temperature_mean.head(10)
```

```python
temp_df = temperature_mean[['cuenca', 'codigo', 'Fecha', 'TMEDIA_CO']]
temp_df.columns = ['basin_id', 'station_id', 'date', 'v_temperature_mean']

temp_df['year'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[0]) , axis = 1) 
temp_df['month'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[1]) , axis = 1) 

temp_df.drop(columns = ['date'], axis = 1, inplace = True)

temp_df = temp_df[['year', 'month', 'basin_id', 'station_id', 'v_temperature_mean']]

temperature_mean = temp_df

print(temperature_mean.dtypes)
temperature_mean.head(10)
```

### Rainfall total

```python
print(rainfall_total.dtypes)
rainfall_total.head(10)
```

```python
temp_df = rainfall_total[['CUENCA', 'Fecha', 'PTPM_CON_WA']]
temp_df.columns = ['basin_id', 'date', 'v_rainfall_mean']

temp_df['year'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[0]) , axis = 1) 
temp_df['month'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[1]) , axis = 1) 

temp_df.drop(columns = ['date'], axis = 1, inplace = True)

temp_df['station_id'] = None

temp_df = temp_df[['year', 'month', 'basin_id', 'station_id', 'v_rainfall_mean']]

rainfall_total = temp_df

print(rainfall_total.dtypes)
rainfall_total.head(10)
```

### Forest Loss Cover

```python
print(forest_loss_cover.dtypes)
forest_loss_cover.head()
```

```python
temp_df = forest_loss_cover[['year', 'month', 'basin_id', 'loss_cover_histo',
                            'loss_cover_fcl', 'loss_cover_fcl_perc', 'loss_cover_perc_acum']]
temp_df.columns = ['year', 'month', 'basin_id', 'v_loss_cover_histo', 'v_loss_cover_fcl', 
                   'v_loss_cover_fcl_perc', 'v_loss_cover_perc_acum']

temp_df['station_id'] = None

temp_df.columns = ['year', 'month', 'basin_id', 'station_id', 'v_loss_cover_histo', 
                   'v_loss_cover_fcl', 'v_loss_cover_fcl_perc', 'v_loss_cover_perc_acum']

forest_loss_cover = temp_df

print(forest_loss_cover.dtypes)
forest_loss_cover.head(10)
```

### Flow mean

```python
print(flow_mean.dtypes)
flow_mean.head(10)
```

```python
temp_df = flow_mean[['cuenca', 'codigo', 'Fecha', 'Q_MEDIA_D']]
temp_df.columns = ['basin_id', 'station_id', 'date', 'v_flow_mean']

temp_df['year'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[0]) , axis = 1) 
temp_df['month'] = temp_df.apply(lambda row: int(str.split(row.date, sep = '-')[1]) , axis = 1) 

temp_df.drop(columns = ['date'], axis = 1, inplace = True)

temp_df = temp_df[['year', 'month', 'basin_id', 'station_id', 'v_flow_mean']]

flow_mean = temp_df

print(flow_mean.dtypes)
flow_mean.head(10)
```

------

## Consolidate

```python
matrix_consol = flow_mean.copy()

matrix_consol = matrix_consol.merge(rainfall_total, 
                                    left_on = ['year', 'month', 'basin_id'], 
                                    right_on = ['year', 'month', 'basin_id'], 
                                    how = 'inner' )

matrix_consol = matrix_consol.merge(temperature_mean, 
                                    left_on = ['year', 'month', 'basin_id'], 
                                    right_on = ['year', 'month', 'basin_id'], 
                                    how = 'inner' )

matrix_consol = matrix_consol.merge(forest_loss_cover, 
                                    left_on = ['year', 'month', 'basin_id'], 
                                    right_on = ['year', 'month', 'basin_id'], 
                                    how = 'inner' )

matrix_consol.drop(columns = ['station_id_y', 'station_id_x'], axis = 1, inplace = True)

print(matrix_consol.dtypes)
print(matrix_consol.shape)
matrix_consol.head()
```

```python
matrix_consol.to_csv('../data/matrix/matrix_consol.csv.zip', index = False, compression = 'zip')
```
