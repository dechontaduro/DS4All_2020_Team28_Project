## Loss Cover Month Matrix

```python
import pandas as pd

pd.options.mode.chained_assignment = None  # default='warn'

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
basins_org = pd.read_csv('../data/matrix/_Defor_acum x MC.csv', error_bad_lines = False, 
                         sep = ';')
```

### Year Month Sequence 

```python
month_seq = pd.date_range(start = '1/1/2000', end = '31/12/2019', freq = 'M')

month_seq = pd.DataFrame(month_seq)
month_seq.columns = ['date']

month_seq['year'] = pd.DatetimeIndex(month_seq['date']).year
month_seq['month'] = pd.DatetimeIndex(month_seq['date']).month

print(month_seq.dtypes)

month_seq.head()
```

### Variable FCL % ACUM

```python
basins_fcl_p_acum_hz = basins_org[['MC', '2001', '2002', '2003', '2004', '2005', '2006',
                                   '2007', '2008', '2009', '2010', '2011', '2012', '2013',
                                   '2014', '2015', '2016', '2017', '2018', '2019']]

basins_fcl_p_acum_hz['2000'] = 0

basins_fcl_p_acum_hz = basins_fcl_p_acum_hz[['MC', '2000', '2001', '2002', '2003', '2004', 
                                             '2005', '2006', '2007', '2008', '2009', '2010', 
                                             '2011', '2012', '2013', '2014', '2015', '2016', 
                                             '2017', '2018', '2019']]

basins_fcl_p_acum_hz.columns = ['mc', '2000', '2001', '2002', '2003', '2004', '2005', 
                                '2006', '2007', '2008', '2009', '2010', '2011', '2012', 
                                '2013', '2014', '2015', '2016', '2017', '2018', '2019']

basins_fcl_p_acum_hz.head()
```

```python
basins_fcl_p_acum = pd.melt(basins_fcl_p_acum_hz, id_vars = ['mc'], 
                       value_vars = ['2000', '2001', '2002', '2003', '2004', '2005', 
                                     '2006', '2007', '2008', '2009', '2010', '2011', 
                                     '2012', '2013', '2014', '2015', '2016', '2017', 
                                     '2018', '2019']).copy()

basins_fcl_p_acum.columns = ['mc', 'year', 'loss_cover']

print(basins_fcl_p_acum)
```

```python
temp_df = basins_fcl_p_acum

temp_df['mc'] = pd.to_numeric(temp_df['mc'])
temp_df['year'] = pd.to_numeric(temp_df['year'])

temp_df = pd.merge(temp_df, month_seq, how = 'right', on = 'year')

temp_df.sort_values(by = ['mc', 'year', 'month'], inplace = True)

temp_df['month'] = pd.to_numeric(temp_df['month'])
temp_df.loc[temp_df['month'] > 1,'loss_cover'] = np.nan
temp_df.loc[(temp_df['year'] == 2000) & 
            (temp_df['month'] == 1),'loss_cover'] = 0

temp_df['type'] = 'real'
temp_df.loc[temp_df['month'] > 1,'type'] = 'interpolate'

list_basins = temp_df.mc.unique()

# Plotting

temp_df_list = []

for i in list_basins:
    temp_df_basin = temp_df[temp_df['mc'] == i].copy().reset_index()
    
    cond1 = (temp_df_basin['year'] == 2019) & (temp_df_basin['month'] == 1)
    cond2 = (temp_df_basin['year'] == 2018) & (temp_df_basin['month'] == 1)
    cond3 = (temp_df_basin['year'] == 2019) & (temp_df_basin['month'] == 12)
    
    loss_cov_ult_year = float(temp_df_basin.loss_cover[cond1])
    loss_cov_pen_year = float(temp_df_basin.loss_cover[cond2])
        
    porc_ults_years = (loss_cov_ult_year / loss_cov_pen_year) - 1
    loss_cover_dec_ult_year = loss_cov_ult_year * (1 + porc_ults_years) 
    
    #print(type(loss_cov_ult_year))
    #print('mc ', i)
    #print('loss_cov_ult_year ', float(loss_cov_ult_year))
    #print('loss_cov_pen_year ', float(loss_cov_pen_year))
    #print('porc_ults_years ', float(porc_ults_years))
    #print('loss_cov_pen_year ', float(loss_cov_pen_year), '\n')
    
    temp_df_basin.loc[cond3, 'loss_cover'] = loss_cover_dec_ult_year
    temp_df_basin.loc[cond3, 'type'] = 'calculated'
   
    #temp_df_basin.to_csv('../temp/tmp.csv', index = False)
    
    #temp_df_basin.interpolate(method = 'polynomial', order = 3, inplace = True)
    temp_df_basin.interpolate(method = 'linear', inplace = True)
    temp_df_basin.set_index(keys = ['date'], inplace = True, drop = True)
    temp_df_list.append(temp_df_basin)
    
    #main_title = 'Loss cover for macro basin ' + str(i)
    #fig, axes = plt.subplots(figsize=(15,5))
    #temp_df_basin.plot(y = 'loss_cover', title = main_title, color = 'red', ax = axes)
    #plt.show()

temp_df = pd.concat(temp_df_list)
temp_df.to_csv('../temp/tempo.csv')

temp_df = temp_df[['mc', 'year', 'month', 'type', 'loss_cover']]

basins_fcl_p_acum = temp_df.copy()
print(basins_fcl_p_acum.tail(30))
```

```python
basins_fcl_p_acum.to_csv('../temp/FCL316Basins_month.csv', index = False)
```
